```python
def rearr(ss):
  sub_loc=[]
  str2=ss.split('SUBCELLULAR LOCATION:')[-1]

  str3=str2.split(';')

  for i in range(len(str3)):
    str4=str3[i].split('{')[0]
    sub_loc.append(str4)

  return sub_loc
```

```python
def go(str):
  str2=str.split('; ')
  sub_loc2=[]
  for i in range(len(str2)):
    str4=str2[i].split(' [')[0]
    sub_loc2.append(str4)
  return sub_loc2
```

```python
for i in range(len(df2)):

  sc=(df2['Gene Ontology (cellular component)'][i])
  # print(sc)
  if (isinstance(sc, float))==False:
    sc2=go(sc)
    df2['Gene Ontology (cellular component)'][i]=sc2
```
