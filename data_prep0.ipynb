```python
for i in range(len(df1[2000:])):
  i=i+2000
  print(i)
  gene=df1['Gene Names (primary)'][i]
  if isinstance(gene, str):
    # print(gene)
    url='https://www.ncbi.nlm.nih.gov/gene/?term='+gene
    r=requests.get(url)
    if r.status_code == 200:
      soup=BeautifulSoup(r.content,"html.parser")
      if soup is not None:
        s=soup.find('em', text='Homo sapiens')
        if s is not None:
          s1=s.find_previous('td')
          if s1 is not None:
            s2=s1.find_previous('td')
            if s2 is not None:
              id=s2.find('input')['value']
              df1['Gene id'][i]=id

  else:
    df1['Gene id'][i]=np.nan
```
